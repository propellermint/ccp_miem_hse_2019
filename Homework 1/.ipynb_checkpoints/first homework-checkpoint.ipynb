### Roman To Decimal

```python
roman_num = {
            "I": 1,
            "V": 5,
            "X": 10,
            "L": 50,
            "C": 100,
            "D": 500,
            "M": 1000,
        }


def roman_to_decimal(rom):
    """Convert a Roman numeral to decimal.
    
    Parameters
    ----------
    rom : str
        A Roman numeral representing a positive integer.
        
    Returns
    -------
    dec : int
        The result of conversion of `rom` into a decimal system.
    """
    dec = 0
    for k in range(len(rom)):
            
        if k > 0 and roman_num[rom[k]] > roman_num[rom[k-1]]:
            dec += roman_num[rom[k]] - 2 * roman_num[rom[k - 1]]
        else:
            dec += roman_num[rom[k]]
    
    return dec

print(roman_to_decimal("MDCL"))
```

```python
test_pairs = [("IX", 9), ("XI", 11), ("MCCII", 1202), ("MMXVIII", 2018), ("XLIX", 49)]

for rom, dec in test_pairs:
    converted = roman_to_decimal(rom)
    print(converted == dec)
```

### Decimal To Roman

```python
def decimal_to_roman1(dec):
    decimals = []
    for k, m in roman_num.items():
        decimals.append(m)
    roman_num_reverse = {
        1: "I",
        5: "V",
        10: "X",
        50: "L",
        100: "C",
        500: "D",
        1000: "M",
        }

    romans = list(roman_num)
    roman = ""

    def lhborders(num):  # find borders to find digit
        hborder = lborder = 1000
        for i in range(len(decimals) - 1):
            if decimals[i] < num < decimals[i + 1]:
                lborder = decimals[i]
                if hborder < decimals[i]:
                    hborder = decimals[i]
                else:
                    hborder = decimals[i + 1]
        return lborder, hborder

    dec_str = str(dec)
    count = 0  # count of digit

    for i in range(len(dec_str)):

        simvol = dec_str[-i - 1]
        digit = 10 ** count
        simvol_ch = int(simvol) * digit

        if simvol_ch in decimals:
            rez = roman_num_reverse.get(simvol_ch)
            roman += rez
        else:
            lborder, hborder = lhborders(simvol_ch)
            if hborder - simvol_ch > digit:
                difference = int((hborder - simvol_ch) / digit)
                if difference * digit < 3 * digit:
                    difference = int((simvol_ch - lborder) / digit)
                    roman = str(roman_num_reverse.get(lborder)) + difference * str(
                        roman_num_reverse.get(digit)) + roman
                else:
                    difference = int((simvol_ch - lborder) / digit)
                    roman = str(roman_num_reverse.get(lborder)) + difference * str(
                        roman_num_reverse.get(digit)) + roman
            else:
                if hborder == lborder:
                    roman = int(simvol_ch / digit) * str(roman_num_reverse.get(hborder)) + roman
                else:
                    roman = str(roman_num_reverse.get(digit)) + str(roman_num_reverse.get(hborder)) + roman
        count += 1
    return roman
```

### A quadratic equation, $x^2 + b x + c = 0$.

```python
import math 
print("Enter a parameteres for quadratic equation: ")
a = float(input("A parameter: "))
b = float(input("B parameter: "))
c = float(input("C parameter: "))

discr = b**2 - 4 * a * c
print(discr)

if discr > 0:
    x1 = (math.sqrt(discr) + b)/(2*a)
    x2 = (math.sqrt(discr) - b)/(2*a)
    print("x1=", + x1,"x2=", + x2)
elif discr ==0:
    x = (math.sqrt(discr) + b)/(2*a)
    print("x=",x)
else:
    print("Discriminant is less than zero")
```

```python
def solve_quad(b, c):
    discr = b**2 / 4.0 - c

    if discr**0.5 == abs(b/2.0):
        x1 = -b
        x2 = -c / b
        
    else:
        x1 = - b / 2.0 + discr**0.5
        x2 = - b / 2.0 - discr**0.5
        
    print(x1, x2)
    return x1, x2
```

```python
from numpy import allclose
variants = [{'b': 4.0, 'c': 3.0},
            {'b': 2.0, 'c': 1.0},
            {'b': 0.5, 'c': 4.0},
            {'b': 1e10, 'c': 3.0},
            {'b': -1e10, 'c': 4.0},]

for var in variants:
    x1, x2 = solve_quad(**var)
    print(allclose(x1*x2, var['c']))
```
