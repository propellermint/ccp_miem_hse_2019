## Task 3
### A snail curve is defined by the following equation in polar coordinates: $r = a + \cos\phi$. Plot the curve on an $x$-$y$ plane for several values of $a$. (Hint: `np.linspace` creates a linearly spaced array of values)


```python
def snail(a, phi):
    r = a + np.cos(phi)
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y

a = np.linspace(-2, 2, 20)
phi = np.linspace(-2 * np.pi, 2 * np.pi, 100)
plt.figure(figsize=(20, 10))

for n in a:
    x, y = snail(n, phi)
    plt.plot(x, y, label = ('a = ', format(round(n, 5))))
    plt.legend()
plt.title('Snail', fontsize=15)
```

## Task 4
### A *cochleoid* curve is defined by the following equation in polar coordinates: $r = a \dfrac{\sin\phi} {\phi }$. Plot the curve on an $x$-$y$ plane. (This time, use `np.where` to work around division by zero).

```python
def cochleoid(a, phi):
    r = a * (np.sin(phi) / phi)
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y

a = np.linspace(-2, 2, 11)
phi = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
phi = np.where(phi != 0, phi, 1000)  #  to work around division by zero
plt.figure(figsize=(15, 10))

for n in a:
    x, y = cochleoid(n, phi)
    plt.plot(x, y, label = ('a = ', format(round(n, 5))))
    plt.legend()
plt.title('Snail', fontsize=15)
```

## Task 6
### Monte-Carlo Error Propagation
Now, we can try using a Monte-Carlo technique instead. The idea behind Monte-Carlo techniques is to generate many possible solutions using random numbers and using these to look at the overall results. In the above case, you can propagate uncertainties with a Monte-Carlo method by doing the following:

* randomly sample values of M1, M2, and r, 1000000 times, using the means and standard deviations
* compute the gravitational force for each set of values

**M1 = 40×104±0.05×104kg**

**M2=30×104±0.1×104kg**

**r=3.2±0.01 m**

**G=6.67384×10−11 m3 kg−1 s−2**

```python
# part 1
M1 = 40e4
M1_std = 0.05e4
M2 = 30e4
M2_std = 0.1e4
r = 3.2
r_std = 0.01
G = 6.67384e-11

F = G * M1 * M2 / r**2
pm = '\xb1'

#  Using the standard error propagation rules we have:
F_std = ((M1_std * G * M2 / r**2)**2 + (M2_std * G * M1 / r**2)**2 + ((-2) * r_std * G * M1 * M2 / r**3)**2)**0.5
print('F(oldschool) = ', F,  pm,  F_std)
```

```python
# sample random M1, M2, r
M1n = np.random.normal(M1, M1_std, size=1000000) # Normal distribution
M2n = np.random.normal(M2, M2_std, size=1000000)
rn = np.random.normal(r, r_std, size=1000000)

Fn = G * M1n * M2n / rn**2
print('F(with some stat stuff) =', Fn.mean(), pm, Fn.std())
```

```python
import scipy.stats as ss

x = np.arange(0.76, 0.81, 0.0001)
fig, axs = plt.subplots(figsize=(10, 7))
axs.hist(Fn, bins = 500, alpha = 0.4, density = True)
axs.plot(x, ss.norm.pdf(x, F, F_std), c='red')
plt.show()
```

```python
# part 2 bigger errors!!! whoah!
# part 1
M1 = 40e4
M1_std = 2e4
M2 = 30e4
M2_std = 1e4
r = 3.2
r_std = 1
G = 6.67384e-11

Fm = G * M1 * M2 / r**2
pm = '\xb1'

#  Using the standard error propagation rules we have:
Fm_std = ((M1_std * G * M2 / r**2)**2 + (M2_std * G * M1 / r**2)**2 + ((-2) * r_std * G * M1 * M2 / r**3)**2)**0.5
print('F(oldschool) = ', F,  pm,  F_std)
```

```python
# sample random M1, M2, r
M11 = np.random.normal(M1, M1_std, size=1000000) # Normal distribution
M21 = np.random.normal(M2, M2_std, size=1000000)
r1 = np.random.normal(r, r_std, size=1000000)

F1 = G * M11 * M21 / r1**2
print('F(with some stat stuff) =', F1.mean(), pm, F1.std())
```

```python
import scipy.stats as ss

x = np.arange(-1, 3, 0.01)
fig, axs = plt.subplots(figsize=(10, 7))
axs.hist(F1[F1 < 15], bins = 500, alpha = 0.7, density = True)
axs.plot(x, ss.norm.pdf(x, Fm, Fm_std), c='black')
plt.show()
```
