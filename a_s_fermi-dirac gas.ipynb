## Распределение Ферми 

   При достаточно низкой температуре статистика Больцмана становится неприемлимой (средние числа заполнения различных квантовых состояний нельзя считать много меньшми 1).
   
   Эта статистика различна в зависимости от того, какого рода волновыми функциями описывается газ, рассматриваемый как система $N$ одинаковых частиц. Функции могут быть либо антисимметричными, либо симметричными по отношению к перестановкам любой пары частиц, первый сулчай - частицы с полуцелым числом, а второй - частицы с целым спином.
   
   Для системы частиц, описывающейся антисимметричными волновыми функциями, справедлив принцип Паули: в каждом квантовом состоянии может находиться одновременно не более одной частицы. Статистика, основанная на этом принципе, называется ***статистикой Ферми-Дирака***. Применим распределение Гиббса к совокупности всех частиц газа, находящихся вданном квантовом состоянии. Обозначим через $\Omega_k$ термодинамический потенциал этой системы:
   
   \begin{equation}
   \label{1}
   \Omega_k=-T*ln\sum_{n_k}\left({e^\frac{\mu-\varepsilon_k}{T}}\right)^{n_k}
   \end{equation}
   
   Согласно принципу Паули числа заполнения могут принимть только два значения: 0, 1. Поэтому получаем: 
   
   \begin{equation}
   \Omega_k=-T*ln\left(1+{e^\frac{\mu-\varepsilon_k}{T}}\right)
   \end{equation}
   
   Среднее число частиц равно производной от потенциала $\Omega$ по химическому потенциалу $\mu$: 
   
   \begin{equation}
   \overline{n_k}=-\frac{\partial\Omega_k}{\partial\mu}=\frac{1}{e^{\left(\varepsilon_k-\mu\right)/T}+1}
   \end{equation}
   
   Это и есть функция распределения для *ферми-газа*. Все $\overline{n_k} \leq 1$. При $exp[(\mu - \epsilon_k) / T] << 1$ функция (3) начинает подчиняться распределению Больцмана.

## Распределение частиц 
Энергия элементарной частицы сводится к кинетической энергии ее поступательного движения:

\begin{equation}
\varepsilon=\frac{1}{2 m}\left(p_{x}^{2}+p_{y}^{2}+p_{z}^{2}\right),
\end{equation}

Распределение по абсолютной величне импульса:

\begin{equation}
dN_p=\frac{gV{p^2}dp}{2\pi^2\hbar\left(e^{\left(\varepsilon_k-\mu\right)/T}+1\right)}, 
\end{equation}

где $g = 2$ - параметр вырождения, так как нет магнитного поля и спин-орбитального взаимодействия. Используя связь энергии с импульсом $\varepsilon=\frac{p^2}{2m}$, получим распределение по энергии:

\begin{equation}\label{eq:'dN'}
 dN_\varepsilon=\frac{gVm^{3/2}}{\sqrt{2}\pi^2\hbar^2}\frac{\sqrt{\varepsilon}d\varepsilon}{e^{\left(\varepsilon-\mu\right)/T}+1}
\end{equation}

## Полное число частиц
Получается интегрированием последнего выражения (\ref{eq:'dN'}):

\begin{equation}\label{eq:'N'}
N=\frac{gVm^{3/2}}{\sqrt{2}\pi^2\hbar^2}\int_0^{\infty}\frac{\sqrt{\varepsilon}d\varepsilon}{e^{\left(\varepsilon-\mu\right)/T}+1}
\end{equation}

Вводя новую переменную $\dfrac{\epsilon}{T} = z$, выражение (\ref{eq:'N'}) перепишется в виде:

\begin{equation}\label{eq:'n/v'}
\dfrac{N}{V}=\dfrac{g(m T)^{3 / 2}}{\sqrt{2} \pi^{2} \hbar^{3}} \int_{0}^{\infty} \frac{\sqrt{z} d z}{e^{z-(\mu / T)} \pm 1}
\end{equation}

Эта формула неявно определяет химический потенциал газа $\mu$ как функцию температуры $T$ и плотности $\dfrac{N}{V}$

## Зависимость температуры от химического потенциала

Если температура $T = 0$, тогда функция распределения равна 1 для всех состояний, энергия которых меньше значения химического потенциала при $T \rightarrow 0 $. Состояния с энергией меньшей химического потенциала при нулевой температуре $\left(\mu_{0}=\epsilon_{F}\right)$ заполнены, при энергии, большей значения химического потенциала - функция распределения равна 0. Функция распределения имеет ступенчатый характер. В формуле \eqref{eq:'n/v'} верхний предел интегрирования вт таком случае $\epsilon_F = \mu$. Интеграл равен:


\begin{equation}\label{eq:'n/v_ef'}
\frac{N}{V}=\frac{g(m T)^{3 / 2}}{\sqrt{2} \pi^{2} \hbar^{3}} \int_{0}^{\epsilon_F} \sqrt{z} d z
\end{equation}


Отсюда определена энергия Ферми:


\begin{equation}\label{eq:'ef'}
\epsilon_{F}=\dfrac{\hbar^{2}}{2 m}\left(\dfrac{3 \sqrt{2}}{4} \pi^{2} \dfrac{N}{V}\right)^{2 / 3}
\end{equation}

Температура Ферми:


\begin{equation}\label{eq:'tf'}
T_{F}=\dfrac{\hbar^{2}}{2 m k_B}\left(\dfrac{3 \sqrt{2}}{4} \pi^{2} \dfrac{N}{V}\right)^{2 / 3}
\end{equation}

При низких температурах из \eqref{eq:'n/v'} и \eqref{eq:'ef'}:

\begin{equation}
\label{eq:'1'}
\frac{2}{3} \epsilon_{F}^{3 / 2}=T^{\frac{3}{2}} \int_{0}^{\infty} \dfrac{\sqrt{z} dz}{e^{z - \mu(T) / T} + 1}
\end{equation}

Для решения данного интеграла численно необходимо обезразмерить величины:

$$\tilde{\mu} = \dfrac{\mu}{T},$$ 
$$\tilde{T} = \dfrac{T}{T_F}$$

Интеграл \eqref{eq:'1'} есть не что иное, как интеграл Ферми-Дирака

\begin{equation}\label{eq:'fermi integral'}
F_{j}(x)=\frac{1}{\Gamma(j+1)} \int_{0}^{\infty} \frac{t^{j}}{e^{t-x}+1} d t
\end{equation}

Это альтернативное определение для полилогарифма:

\begin{equation}\label{eq:'poly'}
F_{j}(x)=-\mathrm{Li}_{j+1}\left(-e^{x}\right)
\end{equation}

Для начала найдем зависимость температуры от химического потенциала:

\begin{equation}\label{eq:'tvsmu'}
\dfrac{2}{3}\tilde{T}^{-\frac{3}{2}} = \int_{0}^{\infty} \dfrac{\sqrt{z} dz}{e^{z - \mu(T) / T} + 1}
\end{equation}

Перепишем \eqref{eq:'tvsmu'} как:

\begin{equation}
I=\int_{0}^{\infty} \frac{\sqrt{z}}{\exp (z-\bar{\mu})+1} d z=\int_{0}^{\infty} \frac{\sqrt{z}}{1+k \exp (z)} d z,
\end{equation}

где $k=e^{-\tilde{\mu}}$

\begin{equation}\label{eq:'i'}
I=-\frac{\sqrt{\pi}}{2} \operatorname{Li}_{\frac{1}{2}}\left(-\frac{1}{k}\right)=-\frac{\sqrt{\pi}}{2} \operatorname{Li}_{\frac{3}{2}}\left(-e^{\tilde{\mu}}\right)
\end{equation}

Для решения можно использовать численный метод, записав

\begin{equation}\label{eq:'fmu'}
f(\bar{\mu})=\log \left(-\frac{\sqrt{\pi}}{2} \operatorname{Li}_{\frac{3}{2}}\left(-e^{\tilde{\mu}}\right)\right)-\log \left(\frac{2}{3} \bar{T}^{-\frac{3}{2}}\right)
\end{equation}


Из \eqref{eq:'fmu'} зависимость $\tilde{T}(\tilde{\mu})$:

\begin{equation}\label{eq:'tmu'}
\bar{T}(\bar{\mu})=\left(-\frac{3}{4} \sqrt{\pi} L i_{3 / 2}(-\exp (\bar{\mu}))\right)^{-2 / 3}
\end{equation}

```python
def f(l):
    return (-3 / 4 * np.pi**0.5 * mp.fp.polylog(3/2, -mp.exp(l)))**(-3 / 2)


mu = np.arange(-2.5, 1, 0.01)
#print(mu)
T = []
for x in mu:
     T.append(mp.re(f(x)))

        
#print(T)
plt.figure(figsize=(15,10))
plt.grid()
plt.xlabel('$\mu$', fontsize='30')
plt.ylabel('$T(\mu)}$', fontsize='30')
plt.plot(mu, T, 'o')
plt.show()
```

```python
#Определяет m(T) при данной температуре

def naive_root(f, t, x_guess, tolerance, step_size):
 
    steps_taken = 0
 
    while abs(f(x_guess, t)) > tolerance:
        if f(x_guess, t) > 0:
            x_guess -= step_size
        elif f(x_guess, t) < 0:
            x_guess += step_size
        else:
            return x_guess
 
        steps_taken += 1
 
    return x_guess, steps_taken
 
F = lambda mu, T: np.real(-((3 * np.sqrt(np.pi)) / 4) * (mp.polylog(3 / 2, float(-np.exp(mu / T))))) - T**(-3 / 2)
root, steps = naive_root(F, 0.1, x_guess=9.5, tolerance=0.1, step_size=0.01)
print("root is:", root)
print ("steps taken:", steps)
%timeit naive_root(F, 0.1, x_guess=9.5, tolerance=0.1, step_size=0.01)
```

Численно можно найти зависимость $\tilde{\mu}(\tilde{T})$

```python
def inverse_quadratic_interpolation(f, t, x0, x1, x2, max_iter, tolerance):
    steps_taken = 0
    while steps_taken < max_iter and abs(x1 - x0) > tolerance:
        fx0 = f(x0, t)
        fx1 = f(x1, t)
        fx2 = f(x2, t)
        L0 = (x0 * fx1 * fx2) / ((fx0 - fx1) * (fx0 - fx2))
        L1 = (x1 * fx0 * fx2) / ((fx1 - fx0) * (fx1 - fx2))
        L2 = (x2 * fx1 * fx0) / ((fx2 - fx0) * (fx2 - fx1))
        new = L0 + L1 + L2
        x0, x1, x2 = new, x0, x1
        steps_taken += 1
    return x0, steps_taken
 
F = lambda mu, T: np.real(-((3 * np.sqrt(np.pi)) / 4) * np.real(mp.fp.polylog(3 / 2, float(-np.exp(mu)))) - T**(-3 / 2))
 
root, steps = inverse_quadratic_interpolation(F, 0.1, 9.4, 9.5, 9.6, max_iter=20, tolerance=1e-7)
print("root is:", root)
print("steps taken:", steps)
%timeit inverse_quadratic_interpolation(F, 0.1, 9.4, 9.5, 9.6, max_iter=20, tolerance=1e-7)
```

```python
#цикл для построения mu(T):

step = 0.01
T = np.arange(0.1, 2, step)
mu_T = []
for i in T:
    mu_T.append(inverse_quadratic_interpolation(F, i, 9.4 + step, 9.5 + step, 9.6 + step,
                                                 max_iter=200, tolerance=1e-9)[0])

plt.figure(figsize=(15,10))
plt.grid()
plt.xlabel('$\~{T}$', fontsize='30')
ax = plt.gca()
ax.xaxis.set_label_coords(0.45, -0.025)
plt.ylabel('$\mu(\~{T})$', fontsize='30')
plt.plot(T, T * mu_T, label='Numerically solved')
plt.axvline(1, linestyle='--', color='red')
plt.axhline(0, linestyle='--', color='red')
plt.text(1,0, '$T*$', rotation=0, fontsize='30')


#Можно сравнить данную зависимость с приближенной
mu_from_book = lambda T: (1 - np.pi**2 / 12 * (T)**2)
plt.plot(T, mu_from_book(T), label='Approximated analytically')
plt.legend()
plt.show()
#print(mu_T)
```

Необходимо определить, когда химический потенциал меняет свой знак:
Функция пересекает еоль только один раз, поэтому, можно использовать метод дихотомии для отыскания корня, предварительно взглянув на график

```python
def bisection(f, a, b, eps=1e-7):
    
    if f(a)*f(b) >= 0:
        print("Bisection method fails.")
        return None
    a_n = a
    b_n = b
    steps = 0
    while abs(b_n - a_n) > eps:
        m_n = (a_n + b_n) / 2
        f_m_n = f(m_n)
        if f(a_n) * f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n) * f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            print("Found exact solution.")
            return m_n
        else:
            print("Bisection method fails.")
            return None
        steps += 1
    return (a_n + b_n) / 2, steps

T_star = bisection(fm, 0.75, 1.5)[0]
print('T*/T_F = ', T_star, '- Точка смены знака mu')
```

### Зная функцию $\tilde{\mu}(\tilde{T})$, необходимо найти интеграл, определяющий полное число частиц

## Полная энергия 
\begin{equation}\label{eq:'E'}
E=\int_0^{\infty}\varepsilon dN_\varepsilon=\frac{gVm^{3/2}}{\sqrt{2}\pi^2\hbar^2}\int_0^{\infty}\frac{\varepsilon^{3/2}d\varepsilon}{e^{\left(\varepsilon-\mu(\tilde{T})\right)/\tilde{T}}+1}
\end{equation}

Тогда плотность энергии: $\frac{E}{N}$:

\begin{equation}\label{eq:'E/N'}
e \equiv \dfrac{E(\tilde{T})}{N} = \dfrac{\frac{gVm^{3/2}}{\sqrt{2}\pi^2\hbar^2}\int_0^{\infty}\frac{\varepsilon^{3/2}d\varepsilon}{e^{\left(\varepsilon-\mu(\tilde{T})\right)/\tilde{T}}+1}}{\frac{gVm^{3/2}}{\sqrt{2}\pi^2\hbar^2}\int_0^{\infty}\frac{\varepsilon^{1/2}d\varepsilon}{e^{\left(\varepsilon-\mu(\tilde{T})\right)/\tilde{T}}+1}} = 
\dfrac{-\tilde{T}^\frac{5}{2} \Gamma\left( \dfrac{5}{2}\right) \operatorname{Li}_{\frac{5}{2}}\left(-e^{\tilde{\mu}(\tilde{T})}\right)}{-\tilde{T}^\frac{3}{2} \Gamma\left( \dfrac{3}{2}\right) \operatorname{Li}_{\frac{3}{2}}\left(-e^{\tilde{\mu}(\tilde{T})}\right)} =\tilde{T}
\dfrac{\Gamma\left( \dfrac{5}{2}\right) \operatorname{Li}_{\frac{5}{2}}\left(-e^{\tilde{\mu}(\tilde{T})}\right)}{\Gamma\left( \dfrac{3}{2}\right) \operatorname{Li}_{\frac{3}{2}}\left(-e^{\tilde{\mu}(\tilde{T})}\right)}
\end{equation}


```python
en = []
tt = np.arange(0.01, 10, 0.001)
print(tt)
def e1(T):
    return (mp.re(T * (mp.gamma(5 / 2) * mp.fp.polylog(5 / 2, - mp.exp(fm(T) / T))) / 
                 (mp.gamma(3 / 2) * mp.fp.polylog(3 / 2, - mp.exp(fm(T) / T)))))

e1(0.01)

for t in tt:
    en.append(e1(t))
    
    
    
def e_from_book(T):
    
    return (mp.re(T * (mp.gamma(5 / 2) * mp.fp.polylog(5 / 2, - mp.exp(mu_from_book(T) / T))) / 
                 (mp.gamma(3 / 2) * mp.fp.polylog(3 / 2, - mp.exp(mu_from_book(T) / T)))))

en_from_book = []

for t in tt:
    en_from_book.append(e_from_book(t))
```

```python
plt.figure(figsize=(15,10))
plt.grid()
plt.title('Density of energy as a function of temperature', fontsize='40')
plt.xlabel('$\~{T}$', fontsize='30')
ax = plt.gca()
ax.xaxis.set_label_coords(0.45, -0.025)
plt.ylabel('$e(\~{T})$', fontsize='30')
plt.plot(tt, en, label='Approximated numerically', color='red')
plt.plot(tt,  3 / 2 * tt, label='Classical', color='red', linestyle='--')
plt.plot(tt, en_from_book, label='Approximated analytically, taken from book', color='darkcyan')
plt.legend()
plt.show()
```
